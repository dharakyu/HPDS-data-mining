```
# TODO: Update Path to the csv file
path = "drive/My Drive/Hybrid Physical+Digital Spaces/Studies/Data Mining/Colab/dictionaries/built-features-vocab-raw.csv"

build_features = pd.read_csv(path)
build_features
```

```
# store the final list of vocab words
lex = []

# add spaces and remove "*"
# this step works in place of Regular Expressions
for feature in build_features:
  for i in range(len(build_features[feature])):
    if f'{build_features[feature][i]}' == 'nan':
      # Skip nan's
      continue
    lex.append(" " + build_features[feature][i])
    if "*" in lex[-1]:
        lex[-1] = lex[-1].replace("*", "")
    else:
        lex[-1] = lex[-1] + " "
        
    lex[-1] = lex[-1].lower()
            
lex

import pandas as pd
df = pd.DataFrame(lex)
df.to_csv('built_features_vocab.csv')
```
